```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
hot_result = []
cold_result = []

for row in data:
    if row[-1] != '':
        if row[-2] != '':
            if int(row[0].split('-')[0]) >= 1983:
                if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                    hot_result.append(float(row[-1]))
                    cold_result.append(float(row[-2]))
plt.plot(hot_result)
plt.plot(cold_result)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan_result = []
octo_result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            octo_result.append(float(row[-1]))
        elif row[0].split('-')[1] == '01':
            jan_result.append(float(row[-1]))
plt.hist(jan_result,bins=100,color='blue')
plt.hist(octo_result,bins=100,color='hotpink')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan_result = []
octo_result = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            octo_result.append(float(row[-1]))
        elif month == '01':
            jan_result.append(float(row[-1]))
plt.hist(jan_result,bins=100,color='blue',label='january')
plt.hist(octo_result,bins=100,color='red',label='august')
plt.legend()
plt.show()
```

```python
#5
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
pay_in = 0
pay_out = 0
unpay_in = 2
unpay_out = 0
for row in data:
    row[4] = int(row[4])
    row[5] = int(row[5])
    row[6] = int(row[6])
    row[7] = int(row[7])
    if row[4] > pay_in:
        pay_in = row[4]
        pay_in_city = row[3]
    if row[5] > pay_out:
        pay_out = row[5]
        pay_out_city = row[3]
    if row[6] > unpay_in:
        unpay_in = row[6]
        unpay_in_city = row[3]
    if row[7] > unpay_out:
        unpay_out = row[7]
        unpay_out_city = row[3]
print("유임승차:", pay_in_city, pay_in)
print("유임하차:",pay_out_city, pay_out)
print("무임승차:",unpay_in_city,unpay_in)
print("무임하차:",unpay_out_city,unpay_out)
```
