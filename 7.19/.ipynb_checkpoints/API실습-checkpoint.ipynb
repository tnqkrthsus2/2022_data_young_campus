```python
url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
serviceKey = 'XwAgJAniwnFcG/5tQWfk3+3+R9aBezCWhvRx/QHl3H/jFWVDuZvtZQoV+T9FevcQ3+NbjZLUk96xdyKnOfK2KA=='
payload ={'serviceKey' : serviceKey , 'pageNo' : '1', 'numOfRows' : '999', 
         'dataType' : 'JSON', 'dataCd' : 'ASOS', 'dateCd' : 'DAY', 'startDt' : '20200101', 
         'endDt' : '20220701', 'stnIds' : '108' }
reqs = requests.get(url, params=payload)
```

```python
load_json = json.loads(reqs.text)
```

```python
load_json.keys()
```

```python
load_json
```

```python
df_weather_days = pd.DataFrame(load_json['response']['body']['items']['item'])
```

```python
df_weather_days
```

```python
df_avgTa = pd.DataFrame(df_weather_days[['tm','avgTa']],copy=True)
```

```python
df_avgTa
```

```python
df_avgTa.set_index('tm',inplace=True)
```

```python
df_avgTa
```

```python
df_avgTa.isna().sum()
```

```python
df_avgTa['avgTa'] = df_avgTa['avgTa'].astype('float64')
```
