```python
#4
import csv
import matplotlib.pyplot as plt
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)
a = input('출력할 호선을 입력하세요')
counts = []
label = []
for row in data:
    row[4:] = map(int, row[4:])
    if row[1] == a:
        counts.append(row[4])
        label.append(row[3])
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(a)
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.bar(label,counts)
plt.show()
```

```python
#5
import csv
import matplotlib.pyplot as plt
f = open('subwayfee.csv')
data = csv.reader(f)
label = next(data)[4:8]
a = input('출력할 호선과 역을 입력하세요')
a_line,a_station = a.split(' ')
c = []
for row in data:
    row[4:] = map(int, row[4:])
    if row[1] == a_line and row[3] == a_station:
        c = row[4:8]
plt.style.use('default')
plt.rc('font', family = 'Malgun Gothic')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.bar(label,c,color='blue')
plt.title(a)
plt.show()
```
